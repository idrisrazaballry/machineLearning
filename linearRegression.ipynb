```python
from sklearn.linear_model import LinearRegression
```

```python
# Step 1
x=df1[['YearsExperience']] # In this fit method we have to selet x as 2-D and y as 1-D
y=df1['Salary']
```

```python
# Step 2
model=LinearRegression()
```

```python
model.fit(x,y)
# Syntax for fit
# modelName.fit(x-train, y-train)  --> It learns the paattern from input and output
# We use this method for model training
```

```python
# Step 3
model.predict([[20]])
# Syntax --> modelName.predict(input)
# The x access only one value i.e. x which must be in 2-D
# This predict method we used in two scenario one is model evaluation and another one future prediction
```

```python
slope=model.coef_
```

```python
slope
```
